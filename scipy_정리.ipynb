# scipy stats

```python
from scipy import stats

# 다른 필요한 모듈들...

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## stats.norm


* norm -> 정규분포

* 넘파이의 np.random.normal과 완전히 겹치는 거 아닌가??

### pdf

```python
# stats.norm.pdf  -> probability density function (정규분포에서의 확률밀도 함수를 의미하는 듯)
## 확률밀도 함수는 Y축 대칭이다.

a1 = np.linspace(-1,1, 100)
a2 = np.linspace(0,1, 100)

plt.subplot(1,2,1)
plt.plot(a1, stats.norm.pdf(a1) )

plt.subplot(1,2,2)
plt.plot(a2, stats.norm.pdf(a2));
```

### rvs

```python
# stats.norm.rvs  -> Random variates of given types

## 정규분포를 기준으로 랜덤한 변수들을 생성한다.
### size는 출력할 변수의 개수
### loc는 변수의 위치, 다시말해서 어떤 값을 중심으로 랜덤하게 변수를 생성할지 결정한다. 
less = stats.norm.rvs(size= 50, loc = 10)
many = stats.norm.rvs(size=1000, loc= 10)

# c의 size를 키워 나갈수록, 분포가 정규분포에 가까워진다.
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(less)
plt.xlabel('Less Random Variable')

plt.subplot(1,2,2)
plt.hist(many)
plt.xlabel('More Ramdom Variable');
```

### 넘파이 활용 (똑같이 구현)

```python
# 넘파이 random.normal 사용해서 똑같이 구현하기
## 중심은 10, 표준편차 3

less = np.random.normal(loc= 10, scale= 3, size= 50 )
more = np.random.normal(loc= 10, scale= 3, size= 1000) 

# 첨부) subplots가 그냥 subplot보다 sharex, sharey를 설정해 줄수 있어서 더 좋은 듯
fig, ax = plt.subplots(1,2,sharex= True)
ax[0].hist(less)
ax[1].hist(more);
```
