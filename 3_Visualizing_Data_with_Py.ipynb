## 전체 학생, 운동학생 데이터 비교하기

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import random
```

```python
#  주어진 데이터, 이걸로 미시간 학생과, 운동하로 가는 학생들의 그래프를 만들 것이다.

mean_uofm = 155
sd_uofm = 5
mean_gym = 185 
sd_gym = 5 
gymperc = .3
totalPopSize = 40000
```

```python
# 위에서 주어진 정보로 정규분포 만들기 -> np.random.normal을 사용하면 된다.

uofm_students = np.random.normal(mean_uofm, sd_uofm, int(totalPopSize* (1-gymperc)) )
students_at_gym = np.random.normal(mean_gym, sd_gym, int(totalPopSize* gymperc))


# 두 데이터를 합치기. np.append사용
population = np.append(uofm_students, students_at_gym)
```

```python
plt.figure(figsize=(10,12))

# Plot the UofM students only
plt.subplot(3,1,1)
sns.distplot(uofm_students)
plt.title("UofM Students Only")
plt.xlim([140,200])

# Plot the Gym Goers only
plt.subplot(3,1,2)
sns.distplot(students_at_gym)
plt.title("Gym Goers Only")
plt.xlim([140,200])

# Plot both groups together
plt.subplot(3,1,3)
sns.distplot(population)
plt.title("Full Population of UofM Students")
plt.axvline(x = np.mean(population))
plt.xlim([140,200])

plt.show()
```

```python
numberSamps = 5000
sampSize = 50

# Get the sampling distribution of the mean from only the gym
mean_distribution = np.empty(numberSamps)

# 5000번 sampling을 반복한다. -> random.chioce사용
for i in range(numberSamps):
    random_students = np.random.choice(population, sampSize)  # -> population에서 sampSize만큼 선택한다.
    mean_distribution[i] = np.mean(random_students)
    
# Plot the population and the biased sampling distribution
plt.figure(figsize = (10,8))

# Plotting the population again
plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full Population of UofM Students")
plt.axvline(x = np.mean(population))
plt.xlim([140,200])  # -> 140에서 200 그래프 영역을 정해주는 군 밑에 그래프 보면 140에서 200사이임

# Plotting the sampling distribution
plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the Mean Weight of All UofM Students")
plt.axvline(x = np.mean(mean_distribution), color = "black")
plt.xlim([140,200])

plt.show()
```

```python
# 헬스장 가는 학생들만을 상대로 전체 데이터 추출해 보기

numberSamps = 5000   # -> sampling을 5000번을 반복한다.
sampSize = 50

# Get the sampling distribution of the mean from only the gym
mean_distribution = np.empty(numberSamps)                                    # -> np.empty는 비어있는 벡터를 만든다고 함.
for i in range(numberSamps):
    random_students = np.random.choice(students_at_gym, sampSize)
    mean_distribution[i] = np.mean(random_students)            # -> 비어 있는 벡터에 평균 값을 추가 한다.
    

# Plot the population and the biased sampling distribution
plt.figure(figsize = (10,8))

# Plotting the population again
plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full Population of UofM Students")
plt.axvline(x = np.mean(population))
plt.xlim([140,200])

# Plotting the sampling distribution
plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the Mean Weight of Gym Goers")
plt.axvline(x = np.mean(students_at_gym), color = "black")
plt.xlim([140,200])

plt.show()
```

####random.sample

```python
# random.sample사용 -> 100사이의 숫자중 5개를 뽑는다.

random.sample(range(100), 5)
```

```python
# 모집단만들고, 샘플링 해서 각각의 평균, 표준편차 계산해보기.

## 모집단 평균은 10, 표준편차는 5, 모집단 수는 10000개
all_number = [random.normalvariate(10, 5) for i in range(10000) ]
```

```python
# 모집단에서 50개를 추출 -> 샘플의 평균과 표준편차는 당연히 모집단과 다르다

a = random.sample(all_number, 50)

print(np.mean(a))
print(np.std(a))
```

```python
# 50개를 추출하는 샘플링을 100번 해보고, 평균과 표준편차를 계산해보자.

sample_mean = [np.mean(random.sample(all_number, 50)) for i in range(100)]
sample_std = [np.std(random.sample(all_number, 50)) for i in range(100)]

print(np.mean(sample_mean))
print(np.mean(sample_std))

# 평균 10과 표준편차 5에 거의 근접했다. -> 셈플 숫자와 추출 횟수를 변경해보며 근접하는 걸 느껴보자
```

## 68-95-99.7 Rule

정규분포.PNG

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

random.seed(1738)
```

```python
mu = 7

sigma = 1.7

Observations = [random.normalvariate(mu, sigma) for _ in range(100000)]
```

```python
# plot을 그리고 표준편차 내에 들어가는 68%의 데이터와, 95%에 들어가는 표준편차2배의 데이터를 볼 수 있다.

sns.distplot(Observations)
plt.axvline(np.mean(Observations)+ np.std(Observations))
plt.axvline(np.mean(Observations)- np.std(Observations))
plt.axvline(np.mean(Observations)+ np.std(Observations)*2, color='y')
plt.axvline(np.mean(Observations)- np.std(Observations)*2, color='y')
```

```python
# Obersevation은 그냥 랜덤한 리스트이다. 
# 판다스가 제공하는 describe를 사용하기 위해선 반드시 그것을 Series로 바꾸어주는 절차가 필요하다

pd.Series(Observations).describe()
```

## distplot 겹쳐 그리기

```python
# 랜덤하게 샘플을 3개 만들었다.

SampleA = random.sample(Observations, 100)
SampleB = random.sample(Observations, 100)
SampleC = random.sample(Observations, 100)
```

```python
# subplots()을 해주면 그래프를 겹쳐서 그릴 수 있다고 한다.
## 리스트처럼 사용해서 도화지에 여러개의 그림을 나눠그릴수 있고, 한 도화지에 여러 그림을 겹쳐서 그릴 수도 있다.
fig, ax = plt.subplots()    # -> 무조건 튜플로 받아줘야 하는 것 같다.

sns.distplot(SampleA, ax = ax)
sns.distplot(SampleB, ax = ax)
sns.distplot(SampleC, ax = ax)

# 완전 랜덤하게 샘플을 뽑았기 때문에 크게 차이가 나지 않는 걸 볼 수 있다.
```
